**First Method: convert the sequence to a list of three nucleoties, find the position of the first ATG and the position of any stop codon, find the number of all possible ORFs and the length of the longest gene then translate it**

```
#convert the sequence into a list of three nucleotides
split_string = [seq[i:i+3] for i in range(0, len(seq), 3)] 
print(split_string, "")
```

```
#function to find how many stop codons
def find_start_and_stop(seq): 
 TAA= seq.count("TAA")
 TAG= seq.count("TAG")
 TGA= seq.count("TGA")
 if TAA > 1: print("There is more than one TAA stop codon. Their counts are ", TAA)
 elif TAA ==1: print("There is only one  TAA Stop codon in this sequence because the Frequency of TAA is", TAA)
 elif TAG > 1: print("There is more than one TAG stop codon. Their counts are ", TAG)
 elif TAG ==1: print("There is only one TAG Stop codon in this sequence because the Frequency of TAG is", TAG)
 elif TGA > 1: print("There is more than one TGA stop codon. Their counts are ", TGA)
 elif TGA ==1: print("There is only one TGA Stop codon in this sequence because the Frequency of TTGA is", TGA)
 elif TAA ==1 and TGA ==TAA and TAG ==1: print("There are three different stop codons")
 elif TAA >1 and TGA >1 and TAG >1: print("There are many different stop codons")
 else: print("There are not any stop codons")
 return
```

```
#testing the function on the list
find_start_and_stop(split_string) 
```

```
#finds the position of first ATG in the list
START_codon=split_string.index("ATG") 
START_codon
```

```
#find positions of all start and stop codons in the list
start_indices = []
for i in range(len(split_string)):
    if split_string[i] == 'ATG':
      start_indices.append(i)

print("ATG position:", start_indices)

stop_indices = []
for i in range(len(split_string)):
    if split_string[i] == 'TAA':
      stop_indices.append(i)

print("TAA position:", stop_indices)
```

```
#find the number of all possible ORFs
#find the the lengths of all possible orfs 
#find the longest ORF
start_indices 
lengths=[]
for i in range(len(start_indices)):
     lengths.append (((7096- start_indices[i])*3)+3)
print("Number of possible ORFs:", len(lengths))
print("POSSIBLE_ORFs_lengths: ", lengths)
print("Longest ORF is", lengths[0])
```

**Second METHOD Modified from https://www.biostars.org/p/384922/**

```
def orf_finder(sequence, frame):
  sequence = sequence.upper()
  for i in range(frame, len(sequence), 3): #note how the step is 3
      codon1 = sequence[i:i+3]
      if codon1 == 'ATG': 
          position_1st_nt_ATG =i
          for j in range(position_1st_nt_ATG, len(sequence), 3):
              codon2 = sequence[j:j+3]
              if codon2 in ['TAA', 'TGA', 'TAG']:
                 position_last_nt_in_Stop= j+3
                 yield (position_last_nt_in_Stop- position_1st_nt_ATG, sequence[position_1st_nt_ATG:position_last_nt_in_Stop]) 
                 break
```

```
from typing import Counter
for orf_length, orf in orf_finder(seq, 0): #first frame means from position 0
    print(orf_length, orf)
```

```
len(list(orf_finder(seq, 0)))
print("There are", len(list(orf_finder(seq, 0))), "possible ORFs" )
```

```
print("The longest ORF:", sorted(list(orf_finder(seq, 0)), reverse=True)[0]) #sort descendingly and show the first one
```

Making a dictionary of codons as keys and what amino acids they code for as values

```
##Taken from https://github.com/justinbois/bootcamp/blob/main/bioinfo_dicts.py
# The set of DNA bases
bases = ['T', 'C', 'A', 'G']

# Build list of codons
codon_list = []
for first_base in bases:
    for second_base in bases:
        for third_base in bases:
            codon_list += [first_base + second_base + third_base]
```

```
##Modified from https://github.com/justinbois/bootcamp/blob/main/bioinfo_dicts.py
# The amino acids that are coded for (* = STOP codon)
amino_acids = 'FFLLSSSSYY**CC*WLLLLPPPPHHQQRRRRIIIMTTTTNNKKSSRRVVVVAAAADDEEGGGG'

# Build dictionary from tuple of 2-tuples (technically an iterator, but it works)
codons = dict(zip(codon_list, amino_acids))
print(codons, "")
```

```
##https://stackoverflow.com/questions/53003785/how-to-extract-only-certain-values-from-a-dictionary-python
proteins = [codons[i] for i in split_string if i in codons] 
print(proteins, "")
```

```
##https://www.geeksforgeeks.org/python-program-to-convert-a-list-to-string/
listToStr = ''.join([str(elem) for elem in proteins])   
print("protein sequence:",listToStr) 
print("length of protein:",len(listToStr)-1) # -1 so we do not count the asterix at the end 
```

**Extracting the Annotation for this gene from the genome**

```
ID = 'NC_045512'
```

```
from Bio import SeqIO
handle = Entrez.efetch(db="nucleotide", id= ID, retmode="text", rettype="gb")
genbank_record = SeqIO.read(handle, "genbank")
```

```
CDSs = [feature for feature in genbank_record.features if feature.type == "CDS"]
len(CDSs)
```
